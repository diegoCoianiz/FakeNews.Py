## Gensim ( representación de documentos como vectores semánticos ):

para modelado de temas no supervisado, indexación de documentos, recuperación por similitud y otras funcionalidades de procesamiento del lenguaje natural.

Gensim es una biblioteca utilizada para el procesamiento de texto y la modelización de temas, diseñada para trabajar de manera eficiente con grandes conjuntos de datos de texto y ofrecer implementaciones de algoritmos populares en procesamiento de lenguaje natural (NLP) y recuperación de información.

**características principales**:

* Implementaciones de **algoritmos de modelado de temas**: Gensim ofrece implementaciones eficientes de algoritmos populares de modelado de temas, como Latent Semantic Analysis (LSA), Latent Dirichlet Allocation (LDA) y Hierarchical Dirichlet Process (HDP). Estos algoritmos son **útiles para descubrir temas latentes en grandes colecciones de texto**.
* Word **Embeddings**: Gensim proporciona herramientas para trabajar con **representaciones vectoriales de palabras, utilizando algoritmos de aprendizaje no supervisado como Word2Vec** y Doc2Vec, capturando la semántica de las palabras y la estructura del documento en un espacio vectorial de alta dimensionalidad.
* Eficiencia en el procesamiento de grandes conjuntos de datos: Gensim está optimizado para trabajar con grandes conjuntos de datos de texto y utiliza técnicas eficientes de procesamiento y almacenamiento para manejar grandes volúmenes de texto de manera escalable.
* **Utilización** de las representaciones vectoriales: Las representaciones vectoriales resultantes se pueden utilizar **en una variedad de tareas de procesamiento de lenguaje natural, como clasificación de texto, clustering de documentos, análisis de sentimientos, recuperación de información y más**. Los modelos de aprendizaje automático pueden tomar estas representaciones como entrada para realizar predicciones o análisis sobre el texto.


## Word Embeddings ( incrustaciones de palabras ):

representaciones vectoriales (matemáticas) de datos donde las distancias lineales capturan la estructura en los conjuntos de datos originales.


La vectorización de palabras en un documento es un proceso esencial en el procesamiento de lenguaje natural (NLP) que implica asignar representaciones vectoriales a cada palabra en el texto. Estas representaciones vectoriales, llamadas incrustaciones o embeddings de palabras, capturan información semántica y contextual sobre las palabras en un espacio de alta dimensionalidad.

* **Preparación** de datos: Antes de vectorizar es importante realizar una limpieza y preprocesamiento del texto. Esto puede incluir tokenización (división del texto en palabras o tokens individuales), eliminación de signos de puntuación, eliminación de palabras vacías (stopwords) y lematización o derivación para normalizar las palabras.
* **Creación del modelo** Word2Vec: Utilizando la biblioteca Gensim, se puede **entrenar un modelo Word2Vec en un corpus** de texto **utilizando una red neuronal** - para aprender representaciones vectoriales de palabras - **que captura la similitud semántica entre palabras basada en su contexto en el corpus** de texto.
* Asignación de vectores de palabras: Una vez entrenado, a cada palabra en el vocabulario del corpus se le asigna a un vector de números reales en el espacio de incrustación. Estos vectores de palabras capturan relaciones semánticas y sintácticas entre las palabras. Por ejemplo, palabras similares estarán más cerca en el espacio vectorial.
* **Vectorización** del documento: Para vectorizar un documento, cada palabra en el documento se representa mediante su vector correspondiente en el espacio de incrustación. **Una vez que se han asignado vectores a todas las palabras del documento, estos vectores pueden agregarse, promediarse u operarse de otras formas para obtener una representación vectorial del documento** en su conjunto.

Estos datos podrían consistir en palabras, en cuyo caso lo llamamos incrustación de palabras. Pero las incrustaciones también pueden representar imágenes, señales de audio e incluso grandes cantidades de datos estructurados.

* **Tokenización** y creación del corpus: Primero, el corpus de texto se tokeniza, lo que significa que se divide en palabras individuales o "tokens". Cada palabra en el corpus se convierte en un elemento de datos individual para el modelo.
* **Creación de pares** de palabras y contexto:pares de palabras y contexto se generan para alimentar al modelo Word2Vec durante el entrenamiento. **Para cada palabra en el corpus se crean pares de palabras de contexto que representan las palabras que aparecen cerca de la palabra objetivo en una ventana de contexto definida**. Por ejemplo, si la palabra objetivo es "gato" y la ventana de contexto es de tamaño 2, los pares de palabras y contexto podrían ser ("gato", "un"), ("gato", "pequeño"), ("gato", "marrón"), etc.
* **Entrenamiento del modelo** Word2Vec: Luego, estos pares de palabras y contexto se utilizan para entrenar el modelo Word2Vec. Durante el entrenamiento, **el modelo ajusta los vectores de palabras de manera que las palabras que comparten contextos similares en el corpus tengan vectores similares en el espacio vectorial**. Esto se logra mediante el ajuste de los pesos de una red neuronal de tipo skip-gram o CBOW (Continuous Bag of Words), que son arquitecturas comunes utilizadas en Word2Vec.
* Optimización de los vectores de palabras: Durante el entrenamiento, el modelo Word2Vec ajusta iterativamente los vectores de palabras para **minimizar una función de pérdida que mide la discrepancia entre las predicciones del modelo y los pares de palabras y contexto reales** en el corpus.
* Generación del vocabulario y asignación de vectores: Después de que el modelo se entrena con el corpus completo, se genera un vocabulario que contiene todas las palabras únicas presentes en el corpus. Cada palabra en el vocabulario se asigna a un vector de números reales en el espacio vectorial, que representa su significado semántico y contexto en el corpus.

```python
# NORMALIZADOR de palabras
def clean_word(word: str) -> str:
  try:
    word=word.lower().strip()
    for letter in word:
      if letter in string.punctuation:
        word=word.replace(letter, "")
    return word
  except:
    print("error: ", word)

clean_word("hello!!! how's it going? ")
```

```python
# NORMALIZADOR de texto
def clean_text(text: str) -> str:
  ct_list=[]
  try:
    for textWord in text.split():
      cleaned_word_=clean_word(textWord)
      if cleaned_word_ not in sw:
        ct_list.append(cleaned_word_)

    return ct_list
  except:
    print("error: ", textWord)


clean_text("hello how old are you? I am 31")
```

```python
""" FUNCIÓN PARA EJECUTAR TODOS LOS MODELOS DE PREDICCIÓN DE FORMA SIMPLE, SIN HIPERPARAMETROS, MEDIR SUS PUNTAJES Y RETORNAR UN DF CON RESULTADOS."""
def basic_modelMetrics_compared(data, model_dict, average='binary'):
    x_tr, x_te, y_tr, y_te = data
    results_list = []  # Lista para almacenar los resultados de cada modelo

    for model_class, model_name in zip(model_dict["model"], model_dict["model_name"]):
        model_instance = model_class()
        fit_model = model_instance.fit(x_tr, y_tr)
        _pred = model_instance.predict(x_te)

        model_results = {"model": model_name}  # Inicializar un diccionario para cada modelo

        for metric_name, metric_function in metrics.items():
            try:
                if 'average' in metric_function.__code__.co_varnames:  # Verificar si 'average' es un argumento válido
                    model_results[f"{metric_name.capitalize()}"] = round(metric_function(y_te, _pred, average=average), 4)
                else:
                    model_results[f"{metric_name.capitalize()}"] = round(metric_function(y_te, _pred), 4)
            except ValueError:  # Manejo de excepciones si la métrica no es aplicable
                model_results[f"{metric_name.capitalize()}"] = "Not Applicable"

        if average == 'binary':
            tn, fp, fn, tp = confusion_matrix(y_te, _pred).ravel()
            # Calcular specificity y manejar la división por cero si es necesario
            specificity_denominator = tn + fp
            if specificity_denominator != 0:
                specificity = round(tn / specificity_denominator, 4)
            else:
                specificity = "Undefined"  # Otra opción sería devolver un valor predeterminado
            model_results["Specificity"] = specificity

        results_list.append(model_results)  # Agregar el diccionario a la lista

    return pd.DataFrame(results_list).sort_values(by=['Accuracy','F1'], ascending=False)  # Convertir la lista de resultados en un DataFrame
```

```python
""" FUNCIÓN PARA EJECUTAR TODOS LOS MODELOS DE PREDICCIÓN CON HIPERPARÁMETROS, MEDIR SUS PUNTAJES Y RETORNAR UN DF CON SUS RESULTADOS"""

def complex_modelMetrics_compared(data, model_dict, param_grid, cv, n_iter, n_jobs, scoring, random_state, verbose, printSteps=False, search="random"):
  # Dividir datos en conjunto de entrenamiento y prueba
  X_train, X_test, y_train, y_test = data
  # Lista para almacenar los resultados de cada modelo
  results_list = []

  # Ciclo de modelos
  for model_class, model_name in zip(model_dict["model"], model_dict["model_name"]):
      print("\t Modelo: ", model_name)
      model  = model_class()

      # Búsqueda de hiperparámetros
      if search=="random":
        grid_search = RandomizedSearchCV(estimator=model, param_distributions=param_grid[model_name],
                                       cv=cv, n_iter=n_iter, n_jobs=n_jobs, scoring=scoring, random_state=random_state, verbose=verbose)
      elif search=="exhaustive":
        grid_search = GridSearchCV(estimator=model, param_grid=param_grid[model_name], cv=cv)
      else:
        print("grid_search must be 'random' for RandomizedSearchCV or 'exhaustive' for GridSearchCV")

      grid_search.fit(X_train, y_train)

      # Selección de mejores hiperparámetros
      best_params = grid_search.best_params_
      if printSteps:
        print("Mejores hiperparámetros encontrados:", best_params)

      # Validación cruzada para evaluar el rendimiento del modelo con los mejores hiperparámetros
      cv_scores = cross_val_score(grid_search.best_estimator_, X_train, y_train, cv=cv, scoring=scoring)
      if printSteps:
        print("Puntajes de validación cruzada:", cv_scores)
        print(f"{scoring} medio de la validación cruzada:", cv_scores.mean())

      # Entrenamiento del modelo final con los mejores hiperparámetros
      best_model = grid_search.best_estimator_
      best_model.fit(X_train, y_train)

      # Evaluación del modelo final en el conjunto de prueba
      test_score = best_model.score(X_test, y_test)
      if printSteps:
        print("Accuracy del modelo final en el conjunto de prueba:", test_score, "\n")

      # Almacenar métricas en un diccionario
      model_results = {
          "Model": model_name,
          "hiperparams": best_params,
          "Accuracy": test_score,
          f"{scoring}": cv_scores
      }
      results_list.append(model_results)  # Agregar el diccionario a la lista

  # Convertir la lista de resultados en un DataFrame
  df_results = pd.DataFrame(results_list)
  print("_"*15,"\n")
  return df_results
```

```python
def run_mlflow_experiment(data, experiment_name, run_name, model, model_params, return_modelInstance=False):
    from datetime import datetime
    """
    Función para ejecutar un experimento de MLflow.

    Parameters:
        data (list): Lista que contiene características y etiquetas en el siguiente orden: X_train, X_test, y_train, y_test.
        experiment_name (str): Nombre del experimento en MLflow.
        run_name (str):  Nombre de la ejecución en MLflow.
        model: Modelo de scikit-learn a utilizar.
        model_params (dict): Parámetros del modelo.

    Returns:
        None
    """

    # Desglosar la lista de datos en características y etiquetas
    X_train, X_test, y_train, y_test = data

    # Crear un experimento o recuperar si ya existe
    mlflow.set_experiment(experiment_name)

    # Iniciar el run con el nombre generado, concatenado con la hora actual
    with mlflow.start_run(run_name= run_name + datetime.now().strftime("%Y-%m-%d_%H-%M-%S")):
        # Instanciar el modelo con los parámetros proporcionados
        model_instance = model(**model_params)

        # Entrenar el modelo
        model_instance.fit(X_train, y_train)

        # Hacer predicciones
        y_pred = model_instance.predict(X_test)

        # Calcular y registrar métricas
        metrics = {
            "accuracy": accuracy_score(y_test, y_pred),
            "precision": precision_score(y_test, y_pred),
            "recall": recall_score(y_test, y_pred)
        }

        for metric_name, metric_value in metrics.items():
            mlflow.log_metric(metric_name, metric_value)

        # Registrar parámetros y modelo en MLflow
        mlflow.log_params(model_params)
        mlflow.sklearn.log_model(model_instance, "{}_model".format(model.__name__))

        # Imprimir la ruta del modelo guardado
        print("Modelo guardado en MLflow con éxito.")

    mlflow.end_run()

    if return_modelInstance : return model_instance
```

##### Explicación del codigo:

* Creación del modelo Word2Vec:

      model=gensim.models.Word2Vec(
        sentences=data["clean_text_"],
        vector_size=EMBEDDING_DIM,
        window=5,
        min_count=1
      )

* sentences: El corpus de texto preprocesado (data["clean_text_"]).
* vector_size: El tamaño de los vectores de palabras, que se establece en EMBEDDING_DIM.
* window: El tamaño de la ventana de contexto, que indica cuántas palabras vecinas se consideran en cada dirección al predecir una palabra objetivo. Aquí, se establece en 5.
* min_count: El recuento mínimo de apariciones de una palabra en el corpus para ser considerada en el modelo. Se establece en 1, lo que significa que todas las palabras se incluirán en el modelo, incluso si solo aparecen una vez.

      def vectorize_text(text: list[str]) -> np.ndarray:

* Se define una función llamada vectorize_text que toma una lista de cadenas < lista de strings (list[str]) > como entrada y devuelve un array numpy.
* Se inicializa un vector de ceros llamado text_vector con la dimensión del espacio de incrustación (EMBEDDING_DIM). Este vector almacenará la representación vectorial del texto.

* Iteración:

        word_vector=model.wv[word]
        text_vector+=word_vector

Para cada palabra se obtiene su vector de palabra correspondiente del modelo Word2Vec (word_vector=model.wv[word]). Por ejemplo:

      model.vw["trump"] =
      array([-8.7500727e-01,  3.7854824e+00, ... ,-2.6890740e+00,  5.1649294e+00], // len = 100
      dtype=float32)

Cualquier palabra, ahora, está representada por un array de 100 números flotantes de 32 bits. El vector de palabra se suma al vector de texto (text_vector+=word_vector), acumulando las representaciones de todas las palabras en el texto.

     text_vector/=len(text) #(avg)

* finalmente se obtiene un promedio **normalizador** de todas las sumas. La normalización asegura que la magnitud del vector de texto no esté influenciada por la longitud del texto. Sin la normalización, los textos más largos tendrían vectores de texto más grandes simplemente debido a la suma de más vectores de palabras, lo que podría introducir sesgos indeseados en el modelo. Los vectores de texto normalizados son más fáciles de interpretar y comparar entre diferentes textos. Al tener una escala común, es más sencillo medir la similitud o la distancia entre los vectores de texto de diferentes documentos o fragmentos de texto.

La normalización del vector de texto asegura que la representación vectorial del texto sea independiente de su longitud, lo que facilita su interpretación, comparación y uso en aplicaciones posteriores de procesamiento de lenguaje

```python
nltk.download('stopwords')
stop_words = nltk.corpus.stopwords.words('english')
stop_words = set(stop_words)

EMBEDDING_DIM = 100

# Cargar el modelo de Word2Vec desde el archivo pickle
with open(word2vecModel_URL, "rb") as f:
    word2vec_model = pickle.load(f)


def clean_word(word: str) -> str:
    """Remove punctuation and lowercase a word

    Args:
        word (str): the word to clean

    Returns:
        str: the cleaned word
    """
    word = word.lower()
    word = word.strip()

    for letter in word:
        if letter in string.punctuation:
            word = word.replace(letter, '')

    return word


def clean_text(text: str) -> list[str]:
    """Remove stop words and punctioation from a whole text.

    Args:
        text (str): the text to clean

    Returns:
        list[str]: the cleaned text
    """
    clean_text_list = []
    for word in text.split():
        cleaned_word = clean_word(word)
        if cleaned_word not in stop_words:
            clean_text_list.append(cleaned_word)

    return clean_text_list


def vectorize_text(text: list[str]) -> np.ndarray:
    """Vectorize a text by doing a sumatory of the word vectors

    Args:
        text (str): the text to vectorize

    Returns:
        np.ndarray: the vectorized text
    """
    text_vector = np.zeros(EMBEDDING_DIM, np.float32)
    for word in text:
        try:
            word_vector = word2vec_model.wv[word]
        except KeyError:
            st.warning(f"Word {word} not in vocabulary")
            continue
        text_vector += word_vector  # equivalent to text_vector = text_vector + word_vector

    text_vector/=len(text) #(avg)

    return text_vector


def classify_embedding(embedding: np.ndarray) -> bool:
    """Classify a text by using ML model

    Args:
        embedding (np.ndarray): the vectorized text. Shape (100,)

    Returns:
        bool: True if the text is real, False otherwise
    """
    embedding = np.expand_dims(embedding, axis=0)  # changes from (100,) to (1, 100)

    data = {
        "inputs": embedding.tolist(),
    }
    prediction=winSVCmodel.predict(data["inputs"])

    return bool(prediction[0])
```

## Flask POST API con Ngrok
* Se crea una API REST con método POST directamente al puerto **localhost 5001** con path **/invocations** para llamadas a **model** (la variable que contiene el modelo registrado traido de MLFlow) para clasificar la data vectorizada que llega desde el body en:
      data = request.get_json()

  y devolver como respuesta un codigo de estatus 200 y la resolución de la clasificación:
      
      return str(is_real), 200

```python
import threading
installedLibrariesForConnection=False

def flaskConnectWithNgrok(auth_token, addr="5001"):
  global installedLibrariesForConnection

  if not installedLibrariesForConnection:
    !pip install pyngrok --quiet
    !pip install mlflow --quiet
    from pyngrok import ngrok
    from flask import Flask, request

    installedLibrariesForConnection=True
    print("Requirements installed \n")

    app = Flask(__name__)

    @app.route('/invocations', methods=['POST'])
    def invocations():
        if request.method == 'POST':
            data = request.get_json()  # Obtener los datos JSON enviados en la solicitud POST
            is_real=classify_embedding(data)
            return str(is_real), 200  # Responder con un mensaje y un código de estado 200

    @app.route('/', methods=['GET'])
    def hello_world():
        return "Hello World", 200  # Responder con "Hello World" y un código de estado 200

    # Terminate open tunnels if they exist
    ngrok.kill()
    # Configura ngrok para exponer tu aplicación Flask
    ngrok.set_auth_token(auth_token)
    public_url = ngrok.connect(addr)
    print(public_url)

    # Ejecutar la aplicación Flask en un hilo separado
    thread = threading.Thread(target=app.run, kwargs={'port': int(addr), 'debug': True, 'use_reloader': False})
    thread.start()
```
