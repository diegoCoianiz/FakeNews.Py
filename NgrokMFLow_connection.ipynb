```python
def MLFlowConnectWithNgrok(auth_token, addr="5000"):
  # Terminate open tunels if exits
  ngrok.kill()
  # Setting authtoken
  ngrok.set_auth_token(auth_token)
  # Open an HTTPs tunnel on port 5000 for 'https://localhost:5000'
  NGROK_TUNNEL=ngrok.connect(addr=addr, proto="http", bind_tls=True)
  print("MLFlow Tracking UI: ", NGROK_TUNNEL.public_url)
  !mlflow ui
  print("-----------------")
  !ps aux | grep mlflow
```

* Para la funcion MLFlowConnectWithNgrok() se debe determinar, como parámetro de autenticación, un codigo válido de la plataforma Ngrok para la construcción de un puente entre el puerto de salida de la terminal local y un puerto de público acceso de la plataforma
* seleccionar, para acceder a la ploataforma de uso compartido de MLFlow, la url que aparece por terminal con la leyenda:
